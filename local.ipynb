# 1- Importing the necessary libraries 

```python
import gzip
import pickle as pic
import matplotlib.pyplot as plt
import numpy as np
```

# 2- Reading the simulation file

```python
file_name = 'sample.sims'
with gzip.GzipFile(file_name, 'rb') as f:
    output = pic.load(f)
```

```python
print(output['data'].keys())
```

# 3- Defining the Viewer class

```python
%matplotlib tk
class Viewer():
    def __init__(self, output , label):
        self.data = output['data'][label]
        self.frame_number = 0
        self.max = len(self.data)
    def show(self):
        fig = plt.figure()
        ax = plt.gca()
        plt.imshow(self.data[self.frame_number])
        cid2  = fig.canvas.mpl_connect('key_press_event', self.__onkeypress__)
        ax.set_title(self.frame_number)
        self.ax = ax
        shape = self.data[self.frame_number].shape
        parameters = output['parameters']
        locs, labels= plt.xticks()
        new_xlabels = [0]
        z = parameters['inner_values']
        new_xlabels.extend(z)
        max_x = shape[1]
        n_labels = len(new_xlabels)
        new_locs = np.linspace(0 , max_x, n_labels)
        plt.xticks(new_locs, ["%.2e" % item for item in new_xlabels])
        plt.xlabel(parameters['inner'])

        locs, labels= plt.yticks()
        print(locs , labels)
        new_ylabels = [0]
        z = parameters['outer_values']
        new_ylabels.extend(z)
        max_y= shape[0]
        n_labels = len(new_ylabels)
        new_locs = np.linspace(0 , max_y, n_labels)
        plt.yticks(new_locs, ["%.2e" % item for item in new_ylabels])
        plt.ylabel(parameters['outer'])
        plt.show()
    def __onkeypress__(self, event):
        if event.key == 'e':
            self.frame_number = self.frame_number+1
            if self.frame_number > self.max -1:
                self.frame_number -= self.max-1
        elif event.key == 'w':
            self.frame_number = self.frame_number -1
            if self.frame_number < 0:
                self.frame_number += self.max - 1
            
        self.img = self.data[self.frame_number]
        plt.imshow(self.img)
        self.ax.set_title(self.frame_number)
        plt.show()
```

# 4- Viewing 
for viewing the next frame press "e" for viewing the previous frame press "w"

```python
view = Viewer(output, 'm')
view.show()
```
